```python
import pandas as pd

daily_infect = [ 15,  11,  21,  25,  32, 186, 204, 334, 242, 274, 294, 315, 722,
       453, 594, 536, 640, 672, 557, 489, 358, 351, 330, 548, 582, 474,
       506, 325, 214]

import pmdarima as pm
model = pm.auto_arima(daily_infect, start_p=2, start_q=2, test='kpss', max_p=5, max_q=2, m=1, d=None, seasonal=False, start_P=1, D=None, trace=True, error_action='warn', stepwise=True)
model.summary()
```

```python
model.predict(10)
```

```python
diag = model.plot_diagnostics(figsize=(20,12))
diag.show()
```
